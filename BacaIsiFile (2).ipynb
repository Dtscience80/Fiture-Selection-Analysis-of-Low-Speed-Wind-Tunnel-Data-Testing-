```python
#Script untuk membaca dari 1 file
filepath = '/content/drive/MyDrive/DATA/run148'  
# Pengambilan data
file = open(filepath,encoding='utf-8')
lines = file.readlines()
i=0
k=0
data_list = []
data= ""

for line in lines:
    if k==1:
      if not ("DPN      P0    Q0    V0") in line:
       data=lines[i].strip()
       data_list.append(data)
       print(data)
       data=""
    if (("DPN      P0    Q0    V0") in line) and (k==1):
       k=0
    elif (("DPN      P0    Q0    V0") in line) and (k==0):
       k=1
    
    i=i+1
    if not line:
        break;
    
file.close()

file1 = open("MyFile.txt", "w") 
for element in data_list:
    file1.write(element + "\n")
file1.close()
```

```python
"""
Program Untuk Cek pengambilan data Clean Dan create file
"""
import glob
files = glob.glob('/content/drive/MyDrive/Riset/Data_Clean/*.*', 
                   recursive = True)
data_list = []
for filepath in files:
# Pengambilan data
  file = open(filepath,encoding='utf-8')
  lines = file.readlines()
  i=0 ; k=0
  data= ""

  for line in lines:
    if k==1:
      if not ("DPN      P0    Q0    V0") in line:
       data=lines[i].strip()
       data_list.append(data)
       print(data)
       data=""
    if (("DPN      P0    Q0    V0") in line) and (k==1):
       k=0
    elif (("DPN      P0    Q0    V0") in line) and (k==0):
       k=1
    
    i=i+1
    if not line:
        break;
    
  file.close()

#Taruh output di folder Hasil run
file1 = open("/content/drive/MyDrive/Riset/Hasil_Run/FileClean_Fix.txt", "w") 
for element in data_list:
    file1.write(element + "\n")
file1.close()
  
```

```python
"""
Program Untuk cek data Deflect dan Flap, dan lainya dari seluruh data 
"""

import glob
files = glob.glob('/content/drive/MyDrive/Riset/DATA_ALL/*.*', 
                   recursive = True)
data_list = []
for filepath in files:
# Pengambilan data
  file = open(filepath,encoding='utf-8')
  lines = file.readlines()
  i=0 ; k=0
  data= ""

  for line in lines:
    if k==1:
      if not ("DFLAP         DRUD       DELEV        DAILR") in line:
       data=(lines[i].strip() + filepath)
       data_list.append(data)
       print(data, end =' ')
       print(filepath)
       data=""
       k=0
    if (k==0) and (("DFLAP         DRUD       DELEV        DAILR") in line):
        k=1
      
    i=i+1
    if not line:
        break;
    
  file.close()

#Taruh output di folder Hasil run
file1 = open("/content/drive/MyDrive/Riset/Hasil_Run/CekFileClean_Alldata.txt", "w") 
for element in data_list:
    file1.write(element + "\n")
file1.close()
```
